```python
# Source - https://stackoverflow.com/a/68793055
# Posted by pmadhu
# Retrieved 2026-02-13, License - CC BY-SA 4.0
# Adapted for voursa.com real estate scraping

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.common.exceptions import (
    StaleElementReferenceException,
    TimeoutException,
    NoSuchElementException,
)
import time
```

```python
# Set up headless Chrome browser
chrome_options = Options()
chrome_options.add_argument("--headless")
chrome_options.add_argument("--no-sandbox")
chrome_options.add_argument("--disable-dev-shm-usage")

driver = webdriver.Chrome(options=chrome_options)
driver.implicitly_wait(10)
driver.get(source1)

# Collect initial listings
properties_selenium = driver.find_elements(By.CSS_SELECTOR, "div.flex-grow.p-4.text-left")
print(f"Initial listings loaded: {len(properties_selenium)}")

# Click "View More" button repeatedly to load all listings
try:
    while True:
        view_more_btn = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.XPATH, "//button[contains(text(), 'View More')]"))
        )
        view_more_btn.click()
        time.sleep(3)

        new_properties = driver.find_elements(By.CSS_SELECTOR, "div.flex-grow.p-4.text-left")
        print(f"Listings after click: {len(new_properties)}")

        # Stop if no new listings were loaded
        if len(new_properties) == len(properties_selenium):
            break
        properties_selenium = new_properties
except (TimeoutException, StaleElementReferenceException, NoSuchElementException):
    # No more "View More" button or page finished loading
    pass

# Final count
properties_selenium = driver.find_elements(By.CSS_SELECTOR, "div.flex-grow.p-4.text-left")
print(f"\nTotal listings loaded: {len(properties_selenium)}")
```

```python
# Parse the fully-loaded page with BeautifulSoup for consistent data extraction
page_source = driver.page_source
driver.quit()

soup1 = BeautifulSoup(page_source, 'html.parser')
properties1 = soup1.find_all(name='div', class_='flex-grow p-4 text-left')
print(f"Total properties extracted from source1: {len(properties1)}")
```

```python
def fetch_data(url):
    try:
        response = requests.get(url)
        response.raise_for_status()
        return response.content
    except requests.exceptions.RequestException as e:
        print(f"Error fetching data from {url}: {e}")
        return None
```

```python

source1 = "https://voursa.com/EN/categories/real_estate"
source2 = "http://wassit.info/immobilier.html"
response1 = fetch_data(source1)
response2 = fetch_data(source2)
```

```python
if response1 is not None:
    soup1 = BeautifulSoup(response1, 'html.parser')
    # Extract data from soup1 as needed
    properties1 = soup1.find_all(name='div', class_='flex-grow p-4 text-left')
    print(len(properties1))
else:    print(f"Failed to retrieve data from {source1}")
if response2 is not None:
    soup2 = BeautifulSoup(response2, 'html.parser')
    # Extract data from soup2 as needed
    properties2 = soup2.find_all(name='div', class_='center')
    response2 = fetch_data("http://wassit.info/immobilier/2-3-2.html")
    soup2 = BeautifulSoup(response2, 'html.parser')
    properties2.extend(soup2.find_all(name='div', class_='center'))
    print(len(properties2))
else:    print(f"Failed to retrieve data from {source2}")
```
